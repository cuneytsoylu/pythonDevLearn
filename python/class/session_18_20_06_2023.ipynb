```python
# palindrom kelimeler
```

```python
def palindrom(x):

    return  x==x[::-1]
```

```python
palindrom("nalan")
```

```python
palindrom("ConnectionAbortedError")
```

```python
(lambda name: name==name[::-1])("sedat")
```

```python
(lambda name: name==name[::-1])("ses")
```

```python
(lambda ** x :x ** 2 if x%2==0 else 2*x)
```
